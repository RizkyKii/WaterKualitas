# **Split data**
fitur untuk membagi data traning dan testing

```python
X = df.drop(['Potability'], axis=1)
Y = df['Potability']
X.columns
```

```python
# Bagi dataset menjadi data latih dan data uji
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 100)
```

```python
X_train.shape, X_test.shape
```

```python
Y_train.shape,Y_test.shape
```

Support Vector Machine

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import classification_report
# model = svm.SVC(kernel='linear', C=0.1, random_state=0)
# ovr = OneVsRestClassifier(model)
```

## training data

```python

model = svm.SVC(kernel='rbf', C=1000, gamma=4)
# model = svm.SVC(kernel='linear', C=1000)
# model = svm.SVC(kernel='sigmoid', C=0.1)
# model = svm.SVC(kernel='poly', C=1,degree=5)
ovr = OneVsRestClassifier(model)
ovr.fit(X_train, Y_train)
Y_pred = ovr.predict(X_train)

print(classification_report(Y_train, Y_pred))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_train, Y_pred)

print('Confusion matrix\n\n', cm)

print('\nTrue Positives(TP) = ', cm[0,0])

print('\nTrue Negatives(TN) = ', cm[1,1])

print('\nFalse Positives(FP) = ', cm[0,1])


print('\nFalse Negatives(FN) = ', cm[1,0])
```

```python
correct_predictions = 0
for true_label, pred_label in zip(Y_train, Y_pred):
    if true_label == pred_label:
        correct_predictions += 1

# Menghitung jumlah total data
total_data = len(Y_train)

# Menghitung akurasi
accuracy = correct_predictions / total_data

print("Akurasi:", accuracy*100)
```

```python
from sklearn.metrics import confusion_matrix
labels = ['Tidak Memenuhi Syarat','Memenuhi Syarat']
cm = confusion_matrix(Y_train, Y_pred)
# Menampilkan confusion matrix
print(cm)

plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.colorbar()
tick_marks = np.arange(len(labels))
plt.xticks(tick_marks, labels, rotation=45)
plt.yticks(tick_marks, labels)
plt.xlabel('Prediksi')
plt.ylabel('Aktual')
plt.title('Confusion Matrix')
plt.show()
```
